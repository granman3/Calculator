```python
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
import matplotlib.pyplot as plt
import tkinter as tk
import numpy as np
import math

def add():
    num1 = float(num1_entry.get())
    num2 = float(num2_entry.get())
    result = num1 + num2
    result_label.config(text=str(result))


def subtract():
    num1 = float(num1_entry.get())
    num2 = float(num2_entry.get())
    result = num1 - num2
    result_label.config(text=str(result))


def multiply():
    num1 = float(num1_entry.get())
    num2 = float(num2_entry.get())
    result = num1 * num2
    result_label.config(text=str(result))


def divide():
    num1 = float(num1_entry.get())
    num2 = float(num2_entry.get())
    if num2 == 0:
        result_label.config(text="Error: Division by zero")
    else:
        result = num1 / num2
        result_label.config(text=str(result))


def sqrt():
    num1 = float(num1_entry.get())
    result = np.sqrt(num1)
    result_label.config(text=str(result))


def exp():
    num1 = float(num1_entry.get())
    num2 = float(num2_entry.get())
    result = np.power(num1, num2)
    result_label.config(text=str(result))


def fact():
    try:
        num1 = int(num1_entry.get())
        if num1 < 0:
            result_label.config(text="Error: Factorial of a negative number")
        else:
            result = math.factorial(num1)
            result_label.config(text=str(result))
    except ValueError:
        result_label.config(text="Error: Factorial of a non-integer")


def circle_d():
    global circle
    new_window = tk.Toplevel(root)
    new_window.title("Circle Area")
    new_window.geometry("500x400")
    global rad_entry
    rad_label = tk.Label(new_window, font=label_font, text="Radius:")
    rad_label.pack()

    rad_entry = tk.Entry(new_window)
    rad_entry.pack()

    def circle_comp():
        # Define the radius of the circle
        r = float(rad_entry.get())

        # Generate an array of theta values from 0 to 2*pi
        theta = np.linspace(0, 2*np.pi, 1000)

        # Calculate the corresponding x and y values for each theta
        x = r * np.cos(theta)
        y = r * np.sin(theta)

        # Plot the circle
        fig, ax = plt.subplots()
        ax.plot(x, y)
        ax.set_aspect('equal')
        ax.set_xlim([-r-1, r+1])
        ax.set_ylim([-r-1, r+1])
        ax.set_title('Circle with radius {}'.format(r))
        ax.fill(x, y, 'blue', alpha=0.2)

        # Embed the plot in a Tkinter window
        canvas = FigureCanvasTkAgg(fig, master=circle)
        canvas.draw()
        canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)

        # Calculate the area of the circle and display it in a label
        area = np.pi * r ** 2
        area_label.config(text=f"Area: {area:.2f}")

    def clear_canvas():
        ax.clear()
        area_label.config(text='')
        canvas.draw()

    check_button = tk.Button(new_window, text="\u2713", command=circle_comp)
    check_button.pack()
    clear_button = tk.Button(new_window, text="Clear", command=clear_canvas)
    clear_button.pack()
    area_label = tk.Label(new_window, font=label_font, text='')
    area_label.pack()

    circle = tk.Canvas(new_window, width=300, height=300)
    circle.pack()
```

```python
import tkinter as tk
import tkinter.font as tkfont
import numpy as np
# Create the GUI
root = tk.Tk()
root.title("Calculator")

root.geometry("640x400")

button_font = ("Helvetica", 24)
label_font = ("Arial", 18)

num1_label = tk.Label(root, font=label_font, text="First number:")
num1_label.grid(row=0, column=0)

num1_entry = tk.Entry(root)
num1_entry.grid(row=0, column=1)

num2_label = tk.Label(root, font=label_font, text="Second number:")
num2_label.grid(row=1, column=0)

num2_entry = tk.Entry(root)
num2_entry.grid(row=1, column=1)

add_button = tk.Button(root, text="+",font=button_font, command= add)
add_button.grid(row=2, column=0)

subtract_button = tk.Button(root, text="-", font=button_font, command=subtract)
subtract_button.grid(row=2, column=1)

multiply_button = tk.Button(root, text="*", font=button_font, command=multiply)
multiply_button.grid(row=3, column=0)

divide_button = tk.Button(root, text="/", font=button_font, command=divide)
divide_button.grid(row=3, column=1)

sqrt_button = tk.Button(root, text="\u221a", font=button_font, command=sqrt)
sqrt_button.grid(row=4, column=1)

exp_button = tk.Button(root, text="x^y", font=button_font, command=exp)
exp_button.grid(row=4, column=0)

fact_button = tk.Button(root, text="x!", font=button_font, command=fact)
fact_button.grid(row=5, column=0)

circle_button = tk.Button(root, text="\u25CF Area", font=button_font, command=circle_d)
circle_button.grid(row=5, column=1)

result_label = tk.Label(root, font=label_font, text="")
result_label.grid(row=0, column=2, columnspan=2)


root.mainloop()
```
